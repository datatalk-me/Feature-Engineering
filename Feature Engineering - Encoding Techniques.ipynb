## Types of Encoding

**Nominal Encoding:** which does not have order 

**ex:** Gender,state

**Ordinal Encoding:** which have order

**ex:** rating,rank

## 3.Ordinal Encoding

```python
import pandas as pd
import numpy as np

data = {'Temprature':['hot','cold','warm','very hot','hot','warm','cold','warm','very hot','cold'],
       'color':['red','red','green','green','white','red','yellow','red','green','white'],
       'target':[1,1,0,1,0,0,1,0,0,1]}
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
Temp_dict = {'cold':1,
            'warm':2,
            'hot':3,
            'very hot':4}

df['Temp_ordinal'] = df.Temprature.map(Temp_dict)
df
```

## 4. Count Or Frequency Encoding

```python
import pandas as pd
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head()
```

```python
columns = [1,3,5,6,7,8,9,13]
```

```python
train_set = train_set[columns]
train_set
```

```python
train_set.columns = ['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
train_set['Country'].head(50)
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
country_map = train_set['Country'].value_counts().to_dict()
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

#### Advantages

1. Easy to implement

2.Not increasing feature space

#### Disadvantages

1. it will provide same weight if the frequencies are same.

## 5. Target Guided Ordinal Encoding

1.ordering the labels according to the target.

2.Replace the labels by Joint Probability of being 1 or 0.

```python
import pandas as pd
df = pd.read_csv('titanic.csv',usecols = ['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace=True)
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2={k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head()
```

## 6.Mean Encoding

```python
mean_ordinal = df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```
